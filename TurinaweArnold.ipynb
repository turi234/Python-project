```python
#identify and display the total number of missing values in each column
data.isnull().sum()
```

```python
#Replace missing values in carat,price, x and y columns
data['carat'].fillna(data['carat'].mean())
data['price'].fillna(data['price'].mean())
data['x'].fillna(data['x'].mean())
data['y'].fillna(data['y'].mean())
```

```python
#Replace missing values in cut,colour,clarity,P and PC
for column in ['cut','colour','clarity','P','PC']:
    data[column].fillna(data[column].mode()[0])
data.isnull().sum()
```

```python
#proof for handling all missing values
data.info()
```

```python
#display the mean,median and mode for carat,price,x and y
cols = ['carat','price','x','y']
for col in cols:
    mean_value = data[col].mean()
    mode_value = data[col].mode()[0]
    median_value = data[col].median()
    print(f"Cols: {col}")
    print(f"Mean: {mean_value}")
    print(f"Median: {median_value}")
    print(f"Mode: {mode_value}")
    print()
```

```python
modes = data[["carat","price","x","y"]].mode().iloc[0]
print(f"modes: \n{modes}")
```

```python
#creating a bar plot showing the distribution of different cut qualities
import matplotlib.pyplot as plt
import seaborn as sns

data['cut'].value_counts().plot(kind ='bar')
plt.title("Distribution of cut qualities")
plt.xlabel("cut")
plt.ylabel("Frequency")
plt.show()
```

```python
#creating a countplot to show the distribution of diamond colour grades
sns.countplot(data,x= "colour")
plt.title("Distribution of diamond colours")
plt.xlabel("Colour Grade")
plt.ylabel("Count")
plt.show()
```

```python
#the correlation matrix for carat, depth, price, x and y
corr_matrix = data[['carat','depth','price','x','y']].corr()
corr_matrix
```

```python
#heatmap to visualize the correlation matrix
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm')
plt.show()
```
